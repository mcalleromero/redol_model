```python
from pathlib import Path
import pandas as pd
```

```python
raw_data = Path("../data/raw")
output_data = Path("../data/original")
```

```python
df = pd.read_csv(raw_data / Path("flag.data"), sep=",")
df
```

```python
df.dropna()
```

```python
non_numeric_features = list(set(df.columns) - set(df._get_numeric_data().columns))
```

```python
df.describe({'categorical': non_numeric_features,
             'numeric': df._get_numeric_data().columns
            })
```

```python
name_of_columns = [f'f{str(x)}' for x in range(len(df.columns)-1)]
name_of_columns.append("class")
```

```python
df.columns = name_of_columns
```

```python
df.info()
```

```python
df = df.rename(columns={"religion": "class"})
```

```python
dummies = pd.get_dummies(df[["topleft", "botright"]])
```

```python
df = df.drop(columns=["topleft", "botright"])
```

```python
df = pd.concat([df, dummies], axis=1)
```

```python
dummies = pd.get_dummies(df[["mainhue"]])
df = df.drop(columns=["mainhue"])
df = pd.concat([df, dummies], axis=1)
```

```python
df
```

```python
df.to_csv(output_data / Path("flags.csv"), index=False)
```
